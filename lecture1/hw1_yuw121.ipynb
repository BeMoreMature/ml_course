1. Create a dictionary of score of each team. That is keys:values are
Team1: 4,
Team2: 3,
Team3: 5,
Team4: 2

2. Add to the dictionary ‘Team5’ who has a score ‘5’ 

3. PRINT all the keys of the dictionary using the keys() method

4. Find the length VALUES in the dictionary (use the len function) and print it

5. Get the average score of the teams and print it (use the sum and len functions)


```python
dict={'Team1': 4, 'Team2': 3, 'Team3': 5, 'Team4': 2}

dict.update({'Team5': 5})

print(dict.keys())

print(len(dict.values()))

sum=0
for value in dict.values():
    sum=sum+value
print('average score is ')
print(sum/len(dict.values()))
```

3. Put the array into a data frame, with column labels ‘x’ and ‘y’ 

4. Plot ‘x’ versus ‘y’

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 


x=np.array((0,1,2,3,4,5))
y=np.array((0,1,4,9,16,25))
arr=np.column_stack((x,y))
print(arr)

print(arr.shape)

myDataFrame=pd.DataFrame({'x':x,'y':y})
print(myDataFrame)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')


# np.average(myDataFrame.iloc[:, 1:2])
np.average(myDataFrame[['y']])
```
